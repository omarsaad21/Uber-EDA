### Feature Enigneering 

#### Format Data

```python
# convert all date columns to date types
df['start_date'] = pd.to_datetime(df['start_date'])
df['end_date'] = pd.to_datetime(df['end_date'])
```

### Get the Speed

```python
# create a new column called 'avg_speed' in km/hr
df['avg_speed'] = round((df['km'] / df['duration']) * 60)
```

### How many trips driven in all Months of year
###  How many trips driven in all days of week
###  How many KMS driven in all period of the day

```python
# create new columns hour, month, day
df['hour'] = df['start_date'].dt.hour
df['month'] =df['start_date'].dt.month_name()
df['day'] = df['start_date'].dt.day_name()
```

```python
df.head(3)
```

```python
# create a func that takes one argument (month, day) which is categorical, and return grouped_df and bar plot figure
def nu_trips_dim(dim):    
    df_grouped = df.groupby(dim)['purpose'].count().reset_index().sort_values(by= 'purpose', ascending= False)
    df_grouped = df_grouped.rename(columns= {'purpose': 'count'}).reset_index(drop= True)
    fig = px.bar(df_grouped, x= dim, y= 'count', color= 'count', color_continuous_scale= 'purp')
    return df_grouped, fig
```

```python
# apply func on month and day
month_df, month_fig = nu_trips_dim('month')
day_df, day_fig = nu_trips_dim('day')
```

```python
# create a func to get period from hour Morning(0, 13), Afternoon(13,19), Evening(19,21), Night else
def get_periods(x):
    if x in range(0,13):
        return 'Morning'
    elif x in range(13,19):
        return 'Afternoon'
    elif x in range(19,21):
        return 'Evening'
    else:
        return 'Night'
```

```python
# create a new column called period
df['period'] = df['hour'].apply(get_periods)
```

```python
df.head(2)
```

```python
df_grouped = df.groupby('period')['km'].sum().reset_index().sort_values(by= 'km', ascending= False)
df_grouped.reset_index(drop= True, inplace= True)
df_grouped
```

```python
px.bar(df_grouped, x= 'period', y= 'km', color= 'km', color_continuous_scale= 'purp')
```
